### Artikelspezifische Untersuchungen der gefährlichen .ch-Domains

```python
# Libraries importieren
import pandas as pd
import numpy as np
```

```python
# Daten importieren
data_tp = pd.read_pickle('data/all_data_update.pkl')
```

```python
# Gesamte Anzahl DNS-Anfragen pro Monat ermitteln
total_requests = {}
for key, data in data_tp.iterrows():
    total_requests[key.strftime("%Y-%m")] = data.sum()
```

```python
# Liste an suspekten Domains einlesen:
import csv
with open("data/domains.txt", "r") as f:
  read_domains = list(csv.reader(f))
suspect_domains = []
for domain in read_domains:
    suspect_domains.append(domain[0].lower())
```

```python
# Anzahl DNS-Anfragen auf suspekte Webseiten pro Monat ermitteln
suspect_requests = {}
for key, data in data_tp.iterrows():
    suspect_sum = 0
    for k, requests in data.iteritems():
        if (k in suspect_domains):
            requests = 0 if np.isnan(requests) else requests
            suspect_sum = requests + suspect_sum
    suspect_requests[key.strftime("%Y-%m")] = suspect_sum
```

```python
# Anteil für die beobachteten Monate berechnen_
for key, value in total_requests.items():
    print("{}: {:.4} %".format(key, 100 / value * suspect_requests[key]))
```

```python
## Daten mit der Summe pro Kategorie aufbereiten:

# dazu URL-Kategorien wieder einlesen:
url_categories_df = pd.read_pickle('data/url_categories.pkl')

# und Bibliotheken laden
from datetime import datetime
import numpy as np





categories_data_sum = pd.DataFrame(columns=["Name", "n_urls", "sum_1902", "sum_1901", "sum_1812", "sum_1811", "sum_1810", "sum_1809", "sum_1808", "sum_1807", "sum_1806", "sum_1805", "sum_1804", "sum_1803", "sum_1802", "sum_1801", "sum_1712", "sum_1711", "sum_1710", "sum_1709", "sum_1708", "sum_1707"])
for cat in url_categories_df.columns.values[2:]:
    visits = {
        "sum_1902": 0,
        "sum_1901": 0, 
        "sum_1812": 0,
        "sum_1811": 0,
        "sum_1810": 0,
        "sum_1809": 0,
        "sum_1808": 0,
        "sum_1807": 0,
        "sum_1806": 0,
        "sum_1805": 0,
        "sum_1804": 0,
        "sum_1803": 0,
        "sum_1802": 0,
        "sum_1801": 0,
        "sum_1712": 0,
        "sum_1711": 0,
        "sum_1710": 0,
        "sum_1709": 0,
        "sum_1708": 0,
        "sum_1707": 0
    }

    for website in (url_categories_df.url[url_categories_df[cat] == True]):
        for average in visits:
            visits[average] += 0 if np.isnan(data_tp[website][datetime.strftime(datetime.strptime(average[4:], '%y%m'), "%Y-%m-%d")]) else data_tp[website][datetime.strftime(datetime.strptime(average[4:], '%y%m'), "%Y-%m-%d")]
    visits["Name"] = cat
    visits["n_urls"] = url_categories_df.url[url_categories_df[cat] == True].count()
    categories_data_sum = categories_data_sum.append(visits, ignore_index=True)

```
